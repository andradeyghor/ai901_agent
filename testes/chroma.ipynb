```python
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction
```

```python
# modelo de embedding
embedding_function = SentenceTransformerEmbeddingFunction(
    model_name="intfloat/multilingual-e5-base"
)
```

```python
# banco vetorial
client = chromadb.PersistentClient(path="./chroma_db")
```

```python
collection = client.get_or_create_collection(
    name="questoes",
    embedding_function=embedding_function
)
```

```python
chunks = [
    "Autovalores representam fatores de escala de uma transformação linear.",
    "Transformações lineares preservam soma vetorial.",
    "Matrizes diagonalizáveis possuem base de autovetores."
]
```

```python
collection.add(
    documents=chunks,
    ids=["1", "2", "3"]
)
```

```python
resultado = collection.query(
    query_texts=["questões difíceis sobre autovalores"],
    n_results=2
)
```
